```python
import csv
reader = csv.reader(open('Output.csv', 'rU'),dialect=csv.excel_tab)
i = 0
X = []
Accuracy = []
Precision = []
Recall = []
FScore = []
for row in reader:
    i+=1
    if i==1:
        continue
    else:
        row = row[0].split(',')
        X.append(int(row[0]))
        Precision.append(float(row[1]))
        Recall.append(float(row[2]))
        FScore.append(float(row[3]))
    
```

```python
from matplotlib import pyplot
X1 = X
Y1 = Accuracy

X2 = X
Y2 = Precision

X3 = X
Y3 = Recall

X4 = X
Y4 = FScore

pyplot.plot(X1,Y1,'g',label = 'Accuracy',linewidth = 2)
pyplot.title('GRAPH')
pyplot.xlabel('X Axis')
pyplot.ylabel('Y Axis')
pyplot.legend()
pyplot.show()
```
